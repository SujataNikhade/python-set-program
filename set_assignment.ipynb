```python
## 1.Write a Python program to create a set.
def create_set():
    s=set()
    n=int(input("how many element? "))
    for i in range(n):
        val=input("enter element: ")
        s.add(val)
    return s
create_set()
```

```python
## Write a Python program to iterate over sets.
def iterate_set(s):
    for i in s:
        print(i)
s={"apple", "banana", "mango", "orange"}
iterate_set(s)
```

```python
## Write a Python program to add member(s) to a set.
def add_member(s):
    n=int(input("how many element to add? "))
    for i in range(n):
        val=input("enter element: ")
        s.add(val)
    return s
s={"apple","banana"}
add_member(s)
```

```python
## Write a Python program to remove item(s) from a given set.
def remove_item(s):
    n=int(input("how many element to remove? "))
    for i in range(n):
        val=input("enter word to remove ")
        if val in s:
            s.remove(val)
        else:
            print(val,"not found in the set")
    return s
s={"apple","banana","mango","orange"}
remove_item(s)
```

```python
## 5.Write a Python program to remove an item from a set if it is present in the set.
def remove_word(w):
    val=input("enter the word you want to remove? ")
    if val in w:
        w.remove(val)
    else:
        print(val,"not found in the set")
    return w
w={"apple","mango","banana","orange","kiwi"}
remove_word(w)
```

```python
## 6.Write a Python program to create an intersection of sets.
def intersection_set(a,b):
    result=[]
    for i in a:
        if i in b:
            result.append(i)
    return result
a=["apple","banana","mango"]
b=["mango","apple","kiwi"]
intersection_set(a,b)
```

```python
## 7.Write a Python program to create a union of sets.
def union_set(A,B):
    for i in A:
        B.add(i)
    return B
A={"apple","banana","mango","kiwi"}
B={"mango","orange","promogranate"}
union_set(A,B)
```

```python
# 8. Write a Python program to create set difference.
def create_set_difference(s1,s2):
    res=set()
    for i in s1:
        if i not in s2:
            res.add(i)
    return res
s1={1,2,3,4,5}
s2={3,4,6,7}
create_set_difference(s1,s2)
```

```python
## 9.Write a Python program to create a symmetric difference.
def symmetric_difference(s1,s2):
    result=[]
    lst=list(s1)
    lst1=list(s2)
    for i in lst:
        if i not in lst1:
            result.append(i)
    for i in lst1:
        if i not in lst:
            result.append(i)
    return result
s1={1,2,3,4,5,6,7}
s2={2,3,6,5,7,8,9}
symmetric_difference(s1,s2)
```

```python
# 10.Write a Python program to check if a set is a subset of another set.
def subset(A,B):
    for i in A:
        if i not in B:
            return False
    return True 
A = {1, 6,7,8,9}
B = {1, 2, 3, 4, 5}
subset(A,B)
```

```python
##14. Find Maximum and Minimum Values in a Set
def min_max(Set):
    lst=list(Set)
    mini=maxi=lst[0]
    for i in lst:
        if i <mini:
            mini=i
        if i>maxi:
            maxi=i
    return mini,maxi
Set={5,3,2,6,7,8,9}
min_max(Set)
```

```python
## 15.Write a Python program to find the length of a set.
def length_set(Set):
    count=0
    for i in Set:
        count+=1
    return count
Set={1,2,3,4,5,6,8}
length_set(Set)
```

```python
## 17. Check if Two Sets Have No Elements in Common
def check_two(set1,set2):
    for i in set1:
        for j in set2:
            if i==j:
                return("common element exist")
    print("No common element")
set1={1,2,3}
set2={3,4,5}
check_two(set1,set2)
```

```python
## 18. Check if a Set is a Superset
def check_set_superset(A,B):
    for i in B:
        if i not in A:
            return False
    return True
A = {1, 2, 3, 4, 5}
B = {2, 3, 5}
check_set_superset(A,B)
```

```python
## 19. Find Elements in a Set Not in Another
def element_not_another(A,B):
    result=set()
    for i in A:
        if i not in B:
            result.add(i)
    return result
A={1,2,3,4,5}
B={2,5,6,7,8}
element_not_another(A,B)
```

```python
#20. Remove Intersection of a Second Set with a First Set
def Remove_intersection(set1,set2):
    result=set()
    for i in set1:
        if i not in set2:
            result.add(i)
    return result
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 6, 7}
Remove_intersection(set1,set2)
```

```python
#Remove duplicates using set but keep original order
lst=[1,3,4,2,3,4,5,2,6,7,8,9]
seen=set()
result=[]
for i in lst:
    if i not in seen:
        result.append(i)
        seen.add(i)
result
```

```python
#Find pairs of numbers whose sum is 10
num={2,3,4,5,6,7,8,9,1}
pair=set()
for i in num:
    if (10-i) in num and i<(10-i):          ## i< 10-i avoids duplicate pairs like (9,1)
        pair.add((i,10-i))
pair
```
